# Peptide data cleaning

```python
data = pd.read_csv('Peptides.csv', header = 0)
data.drop(['Grid Index',
           'Spot SD (Washed)',
           'Alignment',
           'Empty',
           'Thrombin',
           'PSA',
#           'StrepTag', 
           'Myc', 
           'FLAG',
           'NA', 
           'Epitope', 
           'Bad Washed',
           'Bad Strep',
           'Association',
           'Dissociation'], axis = 1, inplace = True)

data.head()
```

```python
data = data.dropna()
data = data[ data['Peptide'] != 'X' ]

data.sort(['Peptide Length', 'Peptide'], ascending = False, inplace = True)
data = data.reset_index(drop = True)

data.head()
```

# Group by everything

```python
grouped = data.groupby(['Peptide',
                        'Peptide Length',
                        'StrepTag',
                        'Tag'])
                        

grouped.head()
```
